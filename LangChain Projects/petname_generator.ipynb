```python
from getpass import getpass
from langchain_community.llms import HuggingFaceHub
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import os
```

```python

HUGGINGFACEHUB_API_TOKEN = getpass()
os.environ["HUGGINGFACEHUB_API_TOKEN"] = HUGGINGFACEHUB_API_TOKEN
#place your huggingface API key after running this cell
```

```python
llm = HuggingFaceHub(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    task="text-generation",
    model_kwargs={
        "max_new_tokens": 200,
        "top_k": 30,
        "temperature": 0.7,
        "repetition_penalty": 1.03,
    },
)
```

```python
promt_template=PromptTemplate(
    input_variables=['animal_type','pet_color'],
    template="I have a {animal_type} pet and I want a cool name for it, it is {pet_color} in color. Suggest me five names for my pet.")
```

```python
chain_name=LLMChain(llm=llm,prompt=promt_template)
```

```python
response = chain_name.run({'animal_type': 'dog', 'pet_color': 'white'})

print("response: ",response)
```
